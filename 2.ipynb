```python
import cv2
import matplotlib.pyplot as plt
```

```python
image_path = 'image2.jpg'  # Replace with your image path
image = cv2.imread(image_path)
```

```python
ksize = (25, 25)  # Kernel size (15x15)
sigmaX = 5  # Standard deviation in the X direction, 0 means calculate from kernel size
blurred_image = cv2.GaussianBlur(image, ksize, sigmaX)
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Blurred Image')
plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
```

```python
plt.show()
```

```python
blurred_image_path = 'blurred_image.jpg'  # Replace with your desired save path
cv2.imwrite(blurred_image_path, blurred_image)
```
